```python
#headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36',
    #'Referer':'https://www.104.com.tw/'}

df = []

for page in range(1,5):
    url= f'https://www.104.com.tw/jobs/search/list?ro=0&kwop=7&keyword=%E4%BA%BA%E8%B3%87%E5%AF%A6%E7%BF%92&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=14&asc=0&page={page}&mode=l&jobsource=cmw_redirect&langFlag=0&langStatus=0&recommendJob=1&hotJob=1'
    #print(url)
    resp = rq.get(url, headers=headers)
    ndf = pd.DataFrame(resp.json()['data']['list'])
    df.append(ndf)
    if ndf.shape[0] < 30:
        break

df = pd.concat(df, ignore_index=True)
```
